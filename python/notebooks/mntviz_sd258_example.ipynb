# mntviz SD258 Example
This notebook shows how to load minutiae from `.min`, convert to numpy, render plots, and export SVG assets used in the README.

```python
from pathlib import Path
from mntviz import load_pairs, minutiae_from_min, plot_mnt, plot_mnt_match
```

```python
repo_root = Path('..').resolve().parent
assets_dir = repo_root / 'python' / 'docs' / 'assets'
assets_dir.mkdir(parents=True, exist_ok=True)

latent_img = repo_root / 'examples' / 'sd258' / '000' / 'images' / 'sd258_000_11-00_latent_bad.png'
latent_min = repo_root / 'examples' / 'sd258' / '000' / 'minutiae' / 'sd258_000_11-00_latent_bad.min'
ref_img = repo_root / 'examples' / 'sd258' / '000' / 'images' / 'sd258_000_11-01_template_bad.png'
ref_min = repo_root / 'examples' / 'sd258' / '000' / 'minutiae' / 'sd258_000_11-01_template_bad.min'
match_file = repo_root / 'examples' / 'sd258' / '000' / 'matching' / 'sd258_000_11-00_latent_bad__sd258_000_11-01_template_bad.txt'

latent_arr = minutiae_from_min(latent_min)
ref_arr = minutiae_from_min(ref_min)
pairs = load_pairs(match_file)
pair_color_values = list(range(len(pairs)))

latent_fig = plot_mnt(
    latent_arr,
    background_img=latent_img,
    output_format='jupyter',
    title='SD258 latent with out_stitch minutiae',
    marker_size=2.5,
    segment_length=8.0,
    line_width=1.2,
    color='#00ff88'
)

ref_fig = plot_mnt(
    ref_arr,
    background_img=ref_img,
    output_format='jupyter',
    title='SD258 reference with out_stitch minutiae',
    marker_size=2.0,
    segment_length=7.0,
    line_width=1.0,
    color='#22d3ee'
)

match_fig = plot_mnt_match(
    left_minutiae=latent_arr,
    right_minutiae=ref_arr,
    pairs=pairs,
    left_background_img=latent_img,
    right_background_img=ref_img,
    output_format='jupyter',
    title='SD258 latent/reference matching',
    left_title='latent',
    right_title='reference',
    marker_size=2.2,
    segment_length=7.5,
    line_width=1.1,
    colormap='tab20',
    colormap_values=pair_color_values,
    show_segments=True,
    left_labels=list(range(len(latent_arr))),
    right_labels=list(range(len(ref_arr))),
    show_labels=True,
)

# HTML files — open in browser and click the SVG button to export
plot_mnt(
    latent_arr,
    background_img=latent_img,
    output_format='html',
    output_path=assets_dir / 'sd258_000_latent_interactive.html',
    title='SD258 latent with out_stitch minutiae',
    marker_size=2.5,
    segment_length=8.0,
    line_width=1.2,
    color='#00ff88'
)

plot_mnt(
    ref_arr,
    background_img=ref_img,
    output_format='html',
    output_path=assets_dir / 'sd258_000_reference_interactive.html',
    title='SD258 reference with out_stitch minutiae',
    marker_size=2.0,
    segment_length=7.0,
    line_width=1.0,
    color='#22d3ee'
)

plot_mnt_match(
    left_minutiae=latent_arr,
    right_minutiae=ref_arr,
    pairs=pairs,
    left_background_img=latent_img,
    right_background_img=ref_img,
    output_format='html',
    output_path=assets_dir / 'sd258_000_matching_interactive.html',
    title='SD258 latent/reference matching',
    left_title='latent',
    right_title='reference',
    marker_size=2.2,
    segment_length=7.5,
    line_width=1.1,
    colormap='tab20',
    colormap_values=pair_color_values,
    show_segments=True,
    left_labels=list(range(len(latent_arr))),
    right_labels=list(range(len(ref_arr))),
    show_labels=True,
)

print('Generated assets in:', assets_dir)
print('Loaded pair file:', match_file.name, '-', len(pairs), 'pairs')
print('Using categorical colormap: tab20')
for p in sorted(assets_dir.glob('sd258_000_*')):
    print('-', p.name)
```
